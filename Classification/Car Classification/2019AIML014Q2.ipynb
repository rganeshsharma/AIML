# Importing Libraries and assigning it to pandas dataframe

```python
import pandas as pd
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/car/car.data"
col_names=['Buying_cost','Maintenance_cost','No_of_Doors','Passenger_Capacity','Boot-Space','Safety_Features','Car-Acceptability']
df = pd.read_csv(url,delimiter=',', header=0, index_col=None,names=col_names)
df.head()
```

```python
# Inspecting datatypes of attributes in our dataset 
df.info()
```

Appending Headers to our Attributes for labeling

By checking car.names file we can find 6 Attributes for our dataset:

buying       v-high, high, med, low

maint        v-high, high, med, low

doors        2, 3, 4, 5-more

persons      2, 4, more

lug_boot     small, med, big

safety       low, med, high

Now let us use Label Encoder and transform all our categorical values in dataset to numerical values

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df = df.apply(label_encoder.fit_transform)
```

```python
df.head() # We can confirm that there is no categorical value and have encoded to Numerical values
```

```python
X = df.iloc[:,0:6]
X
```

```python
y = df[['Car-Acceptability']]
y
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
import warnings
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
warnings.filterwarnings('ignore')
```

```python
# Logistic Regression
log_reg = LogisticRegression(random_state=0).fit(X_train, y_train.values.ravel())
```

```python
# Predicting the Test set results
y_pred = log_reg.fit(X_train, y_train.values.ravel()).predict(X_test)
```

# Confusion Matrix and Accuracy 

```python
cm = confusion_matrix(y_test, y_pred)
print('confusion matrix :\n',cm)
```

```python
acc = accuracy_score(y_test, y_pred)
print('accuracy :',acc)
```

```python
cr = classification_report(y_test, y_pred)
print(cr)
```
