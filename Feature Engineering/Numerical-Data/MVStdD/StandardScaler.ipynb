```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
from sklearn.preprocessing import StandardScaler, scale
```

```python
df = pd.read_csv('absenteeism_processed.csv')
df.head()
```

```python
df.boxplot(figsize=(20,10))
plt.show()
```

```python
# Scikit learn scale function helps us move all the data to unit variance and 0 mean
scaled_df = pd.DataFrame()
scaled_df['Transportation expense'] = scale(df['Transportation expense'])
scaled_df['Distance from Residence to Work'] = scale(df['Distance from Residence to Work'])
scaled_df['Absenteeism time in hours'] = scale(df['Absenteeism time in hours'])
scaled_df.describe()
```

```python
scaled_df.boxplot(figsize=(20,10))
plt.show()
```

```python
# We can use scale in the following ways:
scaled_array = scale(df, copy=True, with_mean=True, with_std=True)
```

```python
type(scaled_array)
```

```python
scaled_array.shape
```

```python
#Scaling all features in the dataset using Standard Scaler
scaled_df = pd.DataFrame(scale(df), columns=df.columns)
scaled_df.describe()
```

```python
scaled_df.boxplot(figsize=(20,10))
```

```python
#Plotting with origoinal means
scaled_df = pd.DataFrame(scale(df, with_mean=False), columns=df.columns)
scaled_df.boxplot(figsize=(20,10))
plt.show()
```

```python
std_scaler = StandardScaler(copy=True, with_mean=True, with_std=True)
std_scaler
```

```python
std_scaler.fit(df)
```

```python
scaled_array = std_scaler.transform(df)
```

```python
scaled_array.shape
```

```python
# We can also use
# scaled_array = std_Scaler.fit_transform(df)
```

```python
scaled_df = pd.DataFrame(scaled_array, columns=df.columns)
scaled_df.boxplot(figsize=(20,10))
plt.show()
```
